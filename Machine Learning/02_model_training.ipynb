```python
X = df2.drop('Rented Bike Count', axis='columns')
y = df2['Rented Bike Count']
```

```python
X.select_dtypes(include=["float64","int64"]).columns
#X.select_dtypes(include="object").columns
```

```python
from sklearn.preprocessing import StandardScaler , OneHotEncoder,FunctionTransformer
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split

num_cols=['Hour', 'Temperature(°C)', 'Humidity(%)', 'Wind speed (m/s)',
       'Visibility (10m)', 'Dew point temperature(°C)']
cat_cols=['Date', 'Seasons', 'Holiday', 'Functioning Day']

num_pipeline=Pipeline([
    ("scl", StandardScaler())
])
cat_pipeline=Pipeline([
    ('scl', OneHotEncoder(sparse_output=False,handle_unknown='ignore')) 
])

preprocessing=ColumnTransformer([
    ("Num",num_pipeline,num_cols),
    ("cat",cat_pipeline,cat_cols),
])
# X_arr = preprocessing.fit_transform(X)
# X_tr = pd.DataFrame(X_arr, columns=preprocessing.get_feature_names_out())
# X_tr.head()

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.2
    )

full_pipeline = Pipeline([
	("cleaning", preprocessing),
	("model", RandomForestRegressor())
])


full_pipeline.fit(X_train, y_train)
y_pred = full_pipeline.predict(X_test)
y_pred
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```
